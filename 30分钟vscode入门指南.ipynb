## 三，熟悉vscode界面

vscode 界面相比jupyter lab 会复杂很多。可以根据需要展开或者折叠它们，也可以拖动调整它们的大小。

1，界面的主要结构

- 活动栏（Activity Bar）：位于界面最左侧，提供对各种视图的快速访问，如文件目录、搜索、版本控制等。

- 主侧栏（Side Bar）：紧邻活动栏右侧，当活动栏选择不同功能按钮呈现不同内容。

- 编辑器区域（Editor Area）：中间是编辑器区域，显示打开的文件内容，可进行代码编辑。

- 副侧栏（Secondary Sidebar）: 位于编辑器区域的左侧，很多时候是折叠的。

- 面板区域（Panel）：通常位于界面下方，提供输出、调试控制台等视图。

- 状态栏（Status Bar）：显示当前git所在分支，错误警告数量，语言模式等重要信息。

2，界面的常用按钮

下面是 vscode 界面中最常用的一些按钮和对应的功能。

- 文件目录：显示当前打开的文件和文件夹。可以通过单击文件或文件夹来切换到它们。

- 插件市场：可以根据需要安装各种插件，甚至可以在里面找到刷微博刷力扣的摸鱼神器。

- 运行按钮：点击它或者按住可以运行当前代码。

- 展开面板：点击它可以找到终端写各种bash命令。再次点击会折叠。

- 展开副侧栏：通常AI编程助手会在这个位置。

## 四、设置vscode环境

1. 设置自动保存 

- 打开设置(Cmd/Ctrl + ,)
- 搜索"Files: Auto Save"
- 选择自动保存模式:
afterDelay: 延迟保存
- PS:这个设置非常重要，默认是不开启自动保存的。
- 有可能写了代码忘了保存，直接关闭了vscode白忙乎了半天。




2. 设置中文界面 
- 打开命令面板(Cmd/Ctrl + Shift + P)
- 输入"Configure Display Language"并选择
- 选择"Install additional languages"
- 在扩展商店中搜索并安装"Chinese (Simplified)"
- 重启vscode生效



3. 设置界面主题
- 打开命令面板(Cmd/Ctrl + Shift + P)
- 输入"Theme"
- 选择"Preference: Color Theme""
- 选择你喜欢的界面主题颜色。
- PS: 资深程序员一般喜欢黑色。



4. 选择Python解释器
- Cmd/Ctrl + Shift + P
- 输入"Python: Select Interpreter"
- 选择对应版本






## 四、常用快捷键

### 通用快捷键

- 打开命令面板: Cmd/Ctrl + Shift + P
- 快速打开文件: Cmd/Ctrl + P
- 保存: Cmd/Ctrl + S

- 撤销: Cmd/Ctrl + Z
- 重做: Cmd/Ctrl + Shift + Z

- 查找: Cmd/Ctrl + F
- 替换: Cmd/Ctrl + Shift + H


### 编辑快捷键

- 注释/取消注释: Cmd/Ctrl + /
- 多光标: Alt + 单击
- 移动行: Alt + Up/Down
- 复制行: Shift + Alt + Up/Down




## 五、推荐插件

1. Python (Microsoft官方开发)
- 提供Python语法高亮
- 调试功能
- 代码智能提示


2. Jupyter (Microsoft官方开发)
- 交互式 运行Jupyter Notebook。
- 右上角切换kernel非常方便。
- 快捷键与Jupyter原生一样 Ctrl/Command+Enter执行代码。
- 可以连接远程Jupyter Notebook服务器上的Kernel  


3. GitHub Copilot 
- 免费，登录github账号即可使用
- 强大的Tab智能补全
- 编码遇到问题，直接在IDE左边栏中开聊，无需切换到浏览器中找其它AI。
- IDE右侧copilot edit模式可以开发完整项目，有时候有点卡就是。
