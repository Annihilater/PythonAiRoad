# FireCrawl，抓取网页变成Markdown

没错，这是一个可以抓取网页内容变成markdown文件的工具。

你可以在网页上直接丢他一个url让它给你抓取然后下载结果，也可以用Python调用它的API接口。

它能够抓取单个网页，还能够根据关键词抓取搜索结果网页中的内容。

每个月每个账号可以免费抓取500个网页页面。

如果你想要更多，很简单，给这个产品打钱 或者 注册多个账号（不要说是我教的，逃）~

此外，这个产品也是开源的，你也可以在本地部署一套。


核心代码逻辑：

```python
from firecrawl import FirecrawlApp
API_KEY = "fc-aaexxxx" 
app = FirecrawlApp(api_key=API_KEY)

#根据url抓取单个页面(如博客文章，wiki百科等)
url = 'https://xxx'
result = app.crawl_url(url) 
text = result['markdown']

#根据关键词抓取搜索页面（一般搜索到wiki百科，百度百科等）
result_list = app.search('达芬奇')
text_list = [x['markdown'] for x in result_list]

```

## 二，Python抓取单网页

首先在这个页面获取api_key 

https://www.firecrawl.dev/app/api-keys

然后安装它的Python接口库

```
pip install firecrawl-py
```

````python
from firecrawl import FirecrawlApp
from robot.secret import CRAWL_API_KEY

app = FirecrawlApp(api_key=CRAWL_API_KEY) #免费额度每个月500个页面

# 我们根据一个我的一篇微信文章url，抓取一个试试
url = 'https://mp.weixin.qq.com/s/lQjuVczZS6BCN1MbuShz0Q'
scraped_data = app.scrape_url(url)
````

````python
from IPython.display import Markdown
Markdown(scraped_data['markdown'][:2000]) 
````
